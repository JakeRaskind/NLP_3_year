В качестве корпуса были использованы аннотации к статьям на сайте cyberleninka.ru , для которых были прописаны ключевые слова. По большей части ключевые слова встречались в аннотации хотя бы раз. Размеченные мной ключевые слова по большей части составляли однословные множества, что хорошо дополняет более расширенно-терминологический вариант ключевых слов с сайта

```python
with open('/content/data.txt', 'r') as f:
  corpus = f.read().split('\n')
```

```python
texts = []
key_words = []
for i in range(len(corpus) // 3):
  texts.append(corpus[i*3])
  key_words.append(list(set(corpus[i*3+1].split('\t')) | set(corpus[i*3+2].split('\t'))))
```

```python
# импорты
import RAKE
import nltk
nltk.download('stopwords')
from nltk.corpus import stopwords
```

```python
pip install pymorphy2
```

```python
from pymorphy2 import MorphAnalyzer
from pymorphy2.tokenizers import simple_word_tokenize
```

```python

stop = stopwords.words('russian')
```

```python
rake = RAKE.Rake(stop)
```

```python
m = MorphAnalyzer()
def normalize_text(text):
    lemmas = []
    for t in simple_word_tokenize(text):
        lemmas.append(
            m.parse(t)[0].normal_form
        )
    return ' '.join(lemmas)
```

```python
rake_kw = []
for text in texts:
  cur_kw_list = rake.run(normalize_text(text), maxWords=3)
  rake_kw.append(cur_kw_list)
```

```python
!pip3 install keybert
```

```python
from keybert import KeyBERT
```

```python
kw_model = KeyBERT('clips/mfaq')
```

```python
bert_kw = []
for text in texts:
  cur = kw_model.extract_keywords(normalize_text(text), keyphrase_ngram_range=(1, 3), stop_words=None)
  bert_kw.append(cur)
```

```python
from summa import keywords
```

```python
tr_kw = []
for text in texts:
  cur = keywords.keywords(normalize_text(text), language='russian', additional_stopwords=stop, scores=True)
  tr_kw.append(cur)
```

```python
print(rake_kw[0], bert_kw[0], tr_kw[0], sep='\n')
```

```python
import numpy as np
```

```python
def metrics_count(res, ref):
  ref = set([normalize_text(kw) for kw in ref])
  res = set([i[0] for i in res])
  tp = ref & res
  fp = res - ref
  fn = ref - res
  print('FP:', fp)
  print('FN:', fn)
  prec = len(tp)/(len(tp)+len(fp))
  rec = len(tp)/(len(tp)+len(fn))
  f1 = 2* prec*rec/(prec+rec + 1)
  return [prec, rec, f1]
```

```python
tr_m, bert_m, rake_m =[], [], []
for i, ref in enumerate(key_words):
 rake_m.append(metrics_count(rake_kw[i], ref))
 tr_m.append(metrics_count(tr_kw[i], ref))
 bert_m.append(metrics_count(bert_kw[i], ref))
print('Rake stats:', np.array(rake_m).mean(axis=0))
print('TextRank stats:', np.array(tr_m).mean(axis=0))
print('KeyBert stats:', np.array(bert_m).mean(axis=0))
```

```python
import spacy_udpipe
```

```python
spacy_udpipe.download("ru")
```

```python
nlp = spacy_udpipe.load("ru")
```

```python
new_texts = []
for text in texts:
  new_text = []
  doc = nlp(text)
  for t in doc:
    if t.pos_ == 'NOUN'  and t.dep_ == 'nmod':
        new_text.extend([t.head.lemma_, t.lemma_])
    elif t.pos_ == 'ADJ':
      new_text.extend([t.lemma_, t.head.lemma_])
    elif t.pos_ == 'ADP':
      new_text.extend([t.head.head.lemma_, t.lemma_, t.head.lemma_])
  new_texts.append(' '.join(new_text))
```

```python
rake_kw = []
for text in new_texts:
  cur_kw_list = rake.run(text, maxWords=3)
  rake_kw.append(cur_kw_list)
```

```python
bert_kw = []
for text in new_texts:
  cur = kw_model.extract_keywords(text, keyphrase_ngram_range=(1, 3), stop_words=None)
  bert_kw.append(cur)
```

```python
tr_kw = []
for text in new_texts:
  cur = keywords.keywords(text, language='russian', additional_stopwords=stop, scores=True)
  tr_kw.append(cur)
```

```python
tr_m, bert_m, rake_m =[], [], []
for i, ref in enumerate(key_words):
 rake_m.append(metrics_count(rake_kw[i], ref))
 tr_m.append(metrics_count(tr_kw[i], ref))
 bert_m.append(metrics_count(bert_kw[i], ref))
print('Rake stats:', np.array(rake_m).mean(axis=0))
print('TextRank stats:', np.array(tr_m).mean(axis=0))
print('KeyBert stats:', np.array(bert_m).mean(axis=0))
```

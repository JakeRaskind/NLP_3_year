```python
from pymystem3 import Mystem
from pymorphy2 import MorphAnalyzer
from natasha import Doc, Segmenter, NewsMorphTagger, NewsEmbedding, MorphVocab
from nltk.tokenize import word_tokenize
from nltk.util import ngrams
```

```python
import json
import re
```

```python
def readFile(filename):
    with open(filename, encoding='utf-8') as f:
        text = f.read()
        return text
```

```python
def pymorphyIt(text):
    res  = []
    tokens = word_tokenize(text)
    m  = MorphAnalyzer()
    for word in tokens:
        ana = m.parse(word)[0]
        res.append((ana.word, str(ana.tag.POS)))
    return res
```

```python
def mystemIt_part1(text):
    stem = Mystem()
    stemmed = stem.analyze(text)
    return stemmed


def  mystemIt_part2(stemmed):
    res = []
    for token in stemmed:
        if token.get('analysis', 0) != 0:
            res.append((token['text'], re.search(r'^(\w*)[,=]', token['analysis'][0]['gr']).group(0)[:-1]))
    return res
```

```python
def natashIt(text):
    res = []
    segmenter = Segmenter()
    morph_vocab = MorphVocab()
    emb = NewsEmbedding()
    morph_tagger = NewsMorphTagger(emb)
    doc = Doc(text)
    doc.segment(segmenter)
    doc.tag_morph(morph_tagger)
    for token in doc.tokens:
        res.append((token.text, str(token.pos)))
    return res
```

```python
text = readFile('dif.txt')
pymorph = pymorphyIt(text)
print(type(pymorph[0][1]))
```

```python
stemmed = mystemIt_part1(text)
```

```python
myst = mystemIt_part2(stemmed)
```

```python
natash = natashIt(text)
```

```python
def accuracy_meter(pred, true):
    correct = 0
    glossary = {'NOUN': ['S', 'NOUN', 'PROPN'],
               'ADJ': ['ADJ', 'A', 'ADJF'],
               'ADV': ['ADV', 'ADVB'],
               'VERB': ['VERB', 'V', 'PRTF', 'INFN', 'AUX', 'GRND'],
               'ADP': ['ADP', 'PREP', 'PR'],
               'ADVPRO': ['ADVPRO'],
               'DET': ['APRO', 'DET'],
               'PRON': ['PRON', 'NPRO', 'SPRO'],
               'PRED': ['PRED'],
               'CCONJ': ['CCONJ'],
               'SCONJ': ['SCONJ'],
               'PART': ['PART', 'PRCL'],
               'NUM': ['NUM', 'NUMR'],
               'PARENTH': ['ARENTH']} 
    cur_t = 0
    cur_p = 0
    while cur_t < len(true):
#         print(true[cur_t][0].lower(), pred[cur_p][0].lower())
        if true[cur_t][0].lower().startswith(pred[cur_p][0].lower()):
            if true[cur_t][0].lower() == pred[cur_p][0].lower() and pred[cur_p][1] in glossary[true[cur_t][1]]:
                correct+= 1
            cur_t += 1
        cur_p += 1
    return correct / len(true)
```

```python
tags = readFile('dif_tags.txt').split()
true = [(tags[i], tags[i+1]) for i in range(0, len(tags), 2)]
print('pymorphy accuracy:', accuracy_meter(pymorph, true))
print('natasha accuracy:', accuracy_meter(natash, true))
print('mystem accuracy:', accuracy_meter(myst, true))
```

```python
def chunker(text):
    res = []
    segmenter = Segmenter()
    morph_vocab = MorphVocab()
    emb = NewsEmbedding()
    morph_tagger = NewsMorphTagger(emb)
    doc = Doc(text)
    doc.segment(segmenter)
    doc.tag_morph(morph_tagger)
    for sent in doc.sents:
        for i in range(len(sent.tokens) - 1):
            if sent.tokens[i].text == 'не' and sent.tokens[i+1].pos == 'VERB' or sent.tokens[i].pos == 'ADJ' and sent.tokens[i+1].pos == 'NOUN' or sent.tokens[i].pos == 'VERB' and sent.tokens[i+1].pos == 'ADV':
                res.append(sent.tokens[i].text + ' ' + sent.tokens[i+1].text)
    return res
```
